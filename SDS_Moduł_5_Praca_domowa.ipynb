## Zadanie - 1

Utwórz macierze a także przeprowadź na nich obliczenia. Postępuj wg. kroków:
* ustaw ziarno losowości równe `5` wykorzystując funkcję dostępną w bibliotece `numpy`;
* wygeneruj 2 ciągi liczb całkowitych z zakresu od 1 do 100. Określ liczbę elementów 20 i 32 odpowiednio. Do tego możesz użyć funkcji: `np.random.randint()`;
* wykorzystaj wygenerowane liczby aby stworzyć macierze `5x4` i `4x8` odpowiednio;
* wykonaj mnożenie stworzonych macierzy;
* podnieś wszystkie elementy macierzy wynikowej z poprzedniego kroku do potęgi `2`;
* odejmij wartość minimalną od każdego elementu macierzy z poprzedniego kroku;
* wylicz sumę wszystkich elementów macierzy;

```
import random
import numpy as np
import plotly.express as px
import pandas as pd
random_state = 5
```

```
# Wygenerowanie ciągów:
A = np.random.randint(0, 100, 20)
B = np.random.randint(0, 100, 32)
```

```
# Nadanie kształtu macierzom:
A = A.reshape(5, 4)
B = B.reshape(4, 8)

print('A  =', '\n', A, '\n', '----------------------------', '\n', 'B = ', '\n', B)
```

```
# Mnożenie macierzy:
C = np.dot(A, B)
print('C = ', '\n', C)
```

```
# Podnieś wszystkie elementy macierzy wynikowej do kwadratu:
C2 = C**2
print('C2 = ', '\n', C2)
```

```
# odejmij wartość minimalną od każdego elementu macierzy z poprzedniego kroku
D = C2 - C2.min()
print('D = ', '\n', D)
```

```
# wylicz sumę wszystkich elementów macierzy:
print(D.sum())
```

```
# Pobranie i prezentacja danych:
dane = pd.read_csv('/content/wineQualityReds.csv')
dane.head(10)
```

```
# Podstawowa analiza danych:
dane.describe()
```

```
# Wykresy rozkładu poszczególnych zmiennych:
cechy = ["fixed.acidity",
         "volatile.acidity",
         "citric.acid",
         "residual.sugar",
         "chlorides",
         "free.sulfur.dioxide",
         "total.sulfur.dioxide",
         "density",
         "pH",
         "sulphates",
         "alcohol",
         "quality"]

for i in range(len(cechy)):
  hist = px.histogram(dane, x = cechy[i], title = f'Wykres rozkładu zmiennej - {cechy[i]}')
  hist.show()
```

```
# Jakie są średnie charakterystyki najlepszego wina:
top_wine = dane[dane['quality'] ==  max(dane['quality'].values)]

print('Liczba najlepiej ocenionych win:', {top_wine['pH'].count()})
print('Średni woltaż najlepiej ocenionych win:', {top_wine['alcohol'].mean()})
print('Średnie pH najlepiej ocenionych win:', {top_wine['pH'].mean()})
```

```
# Jakie są średnie charakterystyki najgorszego wina:
bottom_wine = dane[dane['quality'] ==  min(dane['quality'].values)]

print('Liczba najlepiej ocenionych win:', {bottom_wine['pH'].count()})
print('Średni woltaż najlepiej ocenionych win:', {bottom_wine['alcohol'].mean()})
print('Średnie pH najlepiej ocenionych win:', {bottom_wine['pH'].mean()})
```
