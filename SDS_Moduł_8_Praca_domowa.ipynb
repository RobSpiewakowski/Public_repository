## Zadanie - 2

Wykorzystaj platwormę Filmweb.pl do pobrania informacji o najlepszych 25 filmach z rankingu 500 filmów z całego świata [[Źródło Danych](https://bit.ly/3bnWsSF)]. Strona zawiera informacje: pozycja w rankinhu, nazwa filmu po polsku, nazwa filmu po angielsku, rok premiery, średnia ocena, liczba ocen, itd. Zrealizuj kolejne zadania:

>1. Zescrapuj 25 pierwszych filmów z rankingu, pobierając informacje o: pozycja w rankinhu, nazwa filmu po polsku, nazwa filmu po angielsku, rok premiery, średnia ocena, liczba ocen, link do miniaturki, link do podstrony.

```
# Tworzenie żądania GET:
URL = 'https://www.filmweb.pl/ranking/film'
odp = requests.get(URL)
odp
```

```
# Wyświetlenie zawartości i gotowanie zupy:
html = odp.content
zupa = BeautifulSoup(html, 'lxml')
zupa
```

```
# Pobieranie listy filmów:
movies = zupa.find_all('div', {'class' : 'rankingType'})
movies[0]
```

>2. Stwórz DF oraz uzupełnij pobrane linki, jeżeli nie są pobrane w całości.

```
# Pozycja w rankingu:
pozycja = [span.find('span', {'itemprop' : 'position'}).text.strip() for span in movies]
print(pozycja)
# szuka elementu <span>...</span> a w nim tego, co jest pod kluczem "itemprop"

# Nazwa filmu po polsku:
pol_name = [a.find('a', {'itemprop' : 'url'}).text.strip() for a in movies]
print(pol_name)

# Nazwa filmu w oryginale:
ori_name = [p.find('p', {'class' : 'rankingType__originalTitle'}).text.strip() for p in movies]
print(ori_name)

# Rok premiery:
year = [span.find('span', {'itemprop' : 'datePublished'}).text.strip() for span in movies]
print(year)

# Średnia ocena:
mark = [span.find('span', {'itemprop' : 'ratingValue'}).text.strip() for span in movies]
print(mark)

# Liczba ocen:
count = [span.find('span', {'itemprop' : 'ratingCount'}).text.strip(' ocen społeczności').strip(' oceny') for span in movies]
print(count)

# Link do podstrony:
links = [a.find('a').get('href') for a in movies]
from urllib.parse import urljoin
URL1 = 'https://www.filmweb.pl'
page_lnk = [urljoin(URL1, link) for link in links]
print(page_lnk)
```

```
# Tworzenie Data Frame:
Filmweb_best = pd.DataFrame({'Pozycja w rankingu' : pozycja,
                             'Tytuł polski' : pol_name,
                             "Tytuł oryginału" : ori_name,
                             'Rok premiery' : year,
                             'Ocena na Filmweb' : mark,
                             'Liczba ocen' : count,
                             "Link do podstrony" : page_lnk})
Filmweb_best
```

>3. Wykorzystując linki do podstron z listy pobranych filmów, uzupełnij stworzony DF o informacje: opis filmu, informacje o nagrodach, czas trwania filmu, reżyser, scenariusz, gatunek, produkcja.

```
# Cechy do dodania (Tworzymy puste listy, które będą uzupełniane za pomocą pętli):
movie_description = []          # Opis filmu
awards = []                     # Zdobyte nagrody
duration = []                   # Czas trwania
director = []                   # Reżyser
script = []                     # Autor scenariusza
type_mov = []                   # Gatunek
production = []                 # Produkcja

# Utworzenie pętli, która wygrzebie te cechy z każdej podstrony (korzystając z linków do filmów):
for page in page_lnk:                               # Dla każdej strony filmu
  page_response = requests.get(page)                # Scrapujemy stronę filmu
  page_html = page_response.content                 # Wyświetl zawartość
  page_soup = BeautifulSoup(page_html, 'lxml')      # sparsuj (zupa)

  movie_1 = page_soup.find_all('div', attrs = {'class' : 'filmPosterSection__container page__text'})

  # Opis filmu:
  for span in movie_1:
    x1 = span.find('span', {'itemprop' : 'description'}).text.strip()
  movie_description.append(x1)

  # Informacja o zdobytych nagrodach:
  for div in movie_1:
    x2 = div.find('div', {'class' : 'filmPosterSection__awardsHeaderTitle'})
    if x2 is None:
      x2 = div.find('a', {'class' : 'filmInfo__info'}).text.strip('\xa0')
    else:
      x2 = div.find('div', {'class' : 'filmPosterSection__awardsHeaderTitle'}).text.strip('\xa0')
  awards.append(x2)

  # Czas trwania filmu:
  movie_2 = page_soup.find_all('div', attrs = {'class' : 'filmCoverSection__meta'})

  for span in movie_2:
    x3 = span.find('span', {'class' : 'filmCoverSection__filmTime'}).text.strip()
  duration.append(x3)

  # Reżyser:
  for span in movie_1:
    x4 = span.find('a', {'itemprop' : 'director'}).text.strip(' ')
  director.append(x4)

  # Scenariusz:
  for span in movie_1:
    x5 = span.find('a', {'itemprop' : 'creator'}).text.strip(' ')
  script.append(x5)

  # Gatunek:
  for div in movie_1:
    x6 = div.find('div', {'itemprop' : 'genre'}).text.strip()
  type_mov.append(x6)

  # Produkcja:
  for div in movie_1:
    x7 = div.find(text = 'produkcja').next.find('a').text.strip()
  production.append(x7)
```

```
# Dodanie cech do DataFrame:
Filmweb_best["Opis filmu"] = movie_description
Filmweb_best["Nagrody"] = awards
Filmweb_best["Czas trwania"] = duration
Filmweb_best["Reżyser"] = director
Filmweb_best["scenariusz"] = script
Filmweb_best["Gatunek"] = type_mov
Filmweb_best["Produkcja"] = production
Filmweb_best.head(3)
```

>4. Przekształć czas trwania filmu w minuty.

```
# Podmianka wartości 'godz' -> ':' oraz 'min.' -> '':
Filmweb_best["Czas trwania"] = Filmweb_best["Czas trwania"].str.replace(' godz. ', ':')
Filmweb_best["Czas trwania"] = Filmweb_best["Czas trwania"].str.replace(' min.', "")

# Przekształcenie godzin w minuty, tworzenie dwóch kolumn [h], [m]:
mins = Filmweb_best["Czas trwania"].str.split(':', expand = True).astype(int)

Filmweb_best["Czas trwania"] = (mins[0] * 60 + mins[1])
Filmweb_best.head(3)
```

>5. Jaka średnia długość filmów w TOP25?

```
Filmweb_best["Czas trwania"].mean()
```

>6. Jakiej produkcji są najczęściej filmy w TOP25?

```
Filmweb_best['Produkcja'].value_counts()
```

>7. Filmy jakiej produkcji dostawały Oscar'a? Ile razy wg. TOP25?

```
# Odsiewamy filmy, które mają oskary wśród nagród:
oscars = Filmweb_best[Filmweb_best["Nagrody"].str.contains('Oscar')]

# Z tych elementów tworzymy nowy DataFrame z wybranymi kolumnami ze starego:
oscars = oscars.loc[:, ["Pozycja w rankingu", "Tytuł polski", "Nagrody", "Produkcja"]]
oscars
```
