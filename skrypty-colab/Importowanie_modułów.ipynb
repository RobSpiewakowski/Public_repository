```python
# Przykład Pętli - Trójkąt Pascala:
# Zdefiniowanie pierwszego rzędu:
w = 50
h = 5

numbers = [1]
line = ''
for n in numbers:
    line += "%3d " % (n)
print(line.center(w))

# Pętla, która będzie tworzyła kolejne rzędy aż do rzędu o numerze zdefiniowanym przez użytkownika:
for i in range(h):
# Dopisujemy jedynkę na początku rzędu i zadajemy pozycję:
    newNumbers = [1]
    position = 0
# Druga pętla, która będzie tworzyła kolejne elementy w rzędzie:
    while position < len(numbers) -1:
# Następna liczba jest sumą dwóch nad nią:
        newNumbers.append(numbers[position] + numbers[position + 1])
# Przeskakujemy do następnej liczby:
        position += 1
# No i znów dopisujemy jedynkę:
    newNumbers.append(1)
# Zapisujemy nowy utworzony wiersz:
    numbers = newNumbers.copy()
    
    line = ''
    for n in numbers:
        line += "%3d " % (n)
    print(line.center(w))
```

```python
# Ćwiczenia z Udemy
# Gra w wojnę:
colors = ['Hearts','Diamonds','Clubs','Spades']
figures = [
    {'Figure':'Ace','Power':14},
    {'Figure':'King','Power':13},
    {'Figure':'Queen','Power':12},
    {'Figure':'Jack','Power':11},
    {'Figure':'10','Power':10},
    {'Figure':'9','Power':9}]

import random
all_cards = []
for c in colors:
    for f in figures:
        aCard = f.copy()
        aCard['Color'] = c
        all_cards.append(aCard)    
        
random.shuffle(all_cards)

player1 = all_cards[:12]
player2 = all_cards[12:]
print(player1)
print(player2)

x = 1
while len(player1) > 0 and len(player2) > 0 and x < 10001:
    card1 = player1.pop(0)
    card2 = player2.pop(0)

    print("Rozdanie:", x)
    print('Player 1 ', card1["Figure"] + card1["Color"],':', card2["Figure"] + card2["Color"], ' Player 2')
    if card1["Power"] == card2["Power"]:
        player1.append(card1)
        player2.append(card2)
        print('Draw')
    elif card1["Power"] > card2["Power"]:
        player1.append(card1)
        player1.append(card2)
        print('Player 1 wins')
    else:
        player2.append(card1)
        player2.append(card2)
        print('Player 2 wins')
    x += 1
if x == 10000:
    print("DRAW. Enough")    
elif len(player1) > 0:
    print("Player 1 WON!!!")
else:
    print("Player 2 WON!!!")
```
