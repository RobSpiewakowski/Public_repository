```python
# Wczytywanie plików w formacie .txt

# Funkcja OPEN:
# Aby wczytać plik do skryptu, należy skorzystać z funkcji OPEN. Oto składnia:
# open(ścieżka, tryb) - ścieżka to lokalizacja otwieranego pliku na komputerze lub w innym źródle. Tryby są następujące:
# TRYB ODCZYTU - read mode: open(ścieżka, "r")
# TRYB DO ZAPISU - write mode, tworzy nowy plik lub nadpisuje istniejący: open(ścieżka, "w")
# TRYB DOPISU - append mode: open(ścieżka, "a")
# TRYB TWORZENIA - exclusive creation mode, Tworzy nowy plik: open(ścieżka, "x")
# TRYB DYMYŚLNY - text mode: open(ścieżka, "t")
# TRYB BINARNY - binary mode: open(ścieżka, "b")
# TRYB ROZSZERZONY - pozwala jednocześnie na odczyt i zapis: open(ścieżka, "w+")
```

```python
# Tworzenie nowego pliku .txt - zwykły plik notatnika:
# Jako, że takiego pliku jeszcze nie ma, to tworzymy go. Zapisany będzie w domyslnym katalogu.
plik_text = open("text.txt", "w")
plik_text.write("Hello world. \nI'm Robert. \nand I'm a starting Data Scientist :-)")
plik_text.close()
# Pamiętej, by po każdym przejściu zamykać plik.
```

```python
# Wczytywanie pliku .txt w formacie do odczytu - jako listę:
plik_text = open("text.txt", "r")
# Funkcja READLINES umożliwia otwarcie pliku,w formacie listy
print(plik_text.readlines())
plik_text.close()
```

```python
# Tworzenie pliku .txt z zawartością wygenerowaną w pętli:
# Tworzymy plik tekstowy:
zapetlony = open("zapetlony", "w")

# Tworzymy pętlę, która wygeneruje zawartość pliku:
zapetlony.write("Napisz 100 razy, nie będę tworzył plików .txt poza Python.")
for i in range(101):
  zapetlony.write("Nie będę tworzył plików .txt poza Pythonem. Napisałem to już %d razy. \n" %i)
zapetlony.close()
```

```python
# Użycie funkcji LOADTXT() do wczytania pliku .txt:

# Najpierw importujemy bibliotekę numpy, aby móc tworzyć uporządkowane zbiory danych:
import numpy as np

# Otwieramy plik text2.txt i wklepujemy sobie coś:
text2 = open("text2.txt", "w+")
text2.write("1, 2, 3, 4\n 5, 6, 7, 8\n 9, 10, 11, 12\n 13, 14, 15, 16\n 17, 18, 19, 20\n 21, 22, 23, 24")
text2.close()

# Wczytanie pliku za pomocą funkcji LOADTXT(). Ustawienie przecinka, jako separatora:
dane = np.loadtxt("text2.txt", delimiter = ",")
print(dane)
kreski()
# Sprawdzenie wymiarów:
print('Wymiary tabeli:',dane.shape)
# Sprawdzenie liczby elementów:
print('Liczba elementów:',dane.size)
```

```python
# Jeśli chcemy wczytać tylko wybrane kolumny z danymi:
np.loadtxt("text2.txt", delimiter = ",", usecols = (1,3))
```

```python
# Jeśli chcemy wczytać z pominięciem wiersza:
np.loadtxt("text2.txt", delimiter = ",", skiprows = 1)
```

```python
# Wczytywanie plików w formacie .xml
# .xml - extensable markup language.
# Do pracy z plikami .xml potrzebujesz biblioteki etree ElementTree
import xml.etree.ElementTree as et
```

```python
# Wczytanie pliku .xml za pomocą funkcji PARSE():
plik = et.parse("Sprawozdanie finansowe 2019 BBI.xml")
plik
```

```python
# Pokaż "korzeń" pliku .xml:
korzen = plik.getroot()
korzen.tag
```

```python
# Wyciągnięcie czterech pierwszych elementów z pliku .xml:
element = korzen.getchildren()
print(element[1].tag)
kreski()
print(element[2].tag)
kreski()
print(element[3].tag)
kreski()
print(element[4].tag)
```

```python
# Pokaż atrybuty pierwszego elementu pliku .xml:
atrybut = element[1].attrib
print(atrybut)
```

```python
# Wczytywanie plików w formacie .json:
# Przykład: Tworzymy listę typu krotka (Tupple) i przekształcamy ją w ciąg znaków .json:
import json
a = ("Anna", "Joanna", "Magda", "Weronika", "Jozefa", "Katarzyna", "Boguslawa")
print(a)
json_a = json.dumps(a)
print(json_a)
print(type(json_a))
# Ładowanie pliku .json
lista = json.loads(json_a)
print(lista)
```
