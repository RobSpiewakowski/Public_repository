```python
# Łączenie Data Frame:
# Pandas umożliwia łączenie danych na 4 sposoby:
# - Natural Join / how = 'inner' (iloczyn zbiorów), wynikiem jest zestaw zawierający częśc wspólną łączonych zbiorów,
# - Full outer Join / how = 'outer' (suma zbiorów), wynikiem jest zestaw zawierający wszystkie elementy łączonych zbiorów,
# - left outer join / how = 'left' (różnica zbiorów), wynikiem jest zestaw wszystkich elementów pierwszego z łączonych zbiorów,
# - right outer join / how = 'right' (różnica zbiorów), wynikiem jest zestaw wszystkich elementów drugiego z łączonych zbiorów
```

```python
# Domyślny MERGE:
tab_a = pd.DataFrame({"imie": ["Jozef", "Magda", "Svetlana", "Jedrzej"], "wiek": [25, 31, 89, 65], "status_emeryt": [False, False, True, True]})
tab_b = pd.DataFrame({"status_emeryt": [True, False], "znizka": [10, 0]})
print(tab_a)
kreski()
print(tab_b)
kreski()
# Łączenie tabel typu DataFrame - łączenie względem zmiennej 'status_emeryt':
pd.merge(tab_a, tab_b, on = "status_emeryt")

# Funkcja MERGE działa domyślnie jako left outer join, tj. tabela tab_b została "doklejona" do tabeli tab_a, z pominięciem elementu ze wspólną zmienną (status_emeryt).
```

```python
# W ramach funkcji MERGE możemy również łączyć tabele względem zmiennej, która ma w tych tabelach różne nazwy:
# Domyślny MERGE:
tab_c = pd.DataFrame({"imie": ["Zygmunt", "Jagoda", "Angelika", "Bronisław"], "wiek": [63, 25, 75, 36], "status_emeryt": [True, False, True, False]})
tab_d = pd.DataFrame({"emeryt?": [True, False], "znizka": [15, 0]})
print(tab_c)
kreski()
print(tab_d)
kreski()
# Składnia: pandas.merge(pierwszy df, drugi df, left_on = "nazwa kolumny", right_on = "nazwa kolumny odpowiadającej")
pd.merge(tab_c, tab_d, left_on = "status_emeryt", right_on = "emeryt?")
```

```python
# Łączenie tabel typu LEFT JOIN:
tab_e = pd.DataFrame({"imie": ["Adam", "Andrzej", "Anzelm", "Alojzy", "Arkadiusz"], "wiek": [19, 36, 94, 56, 42]})
tab_f = pd.DataFrame({"wiek": [36, 94, 42, 82], "sztuczna_szczeka": [False, True, False, True]})
print(tab_e)
kreski()
print(tab_f)
kreski()
pd.merge(tab_e, tab_f, on = "wiek", how = "left")
# Jak widać, dla wszystkich panów z tabeli 1, którym nie przypisano wartości zmiennej 'sztuczna_szczeka', pojawiło się NAN (nieznane). Nie ma tu również wartości z tabeli tab_f, które nie mają odpowiednika w tabeli tab_e.
```

```python
# Łączenie tabel typu RIGHT JOIN:
tab_e = pd.DataFrame({"imie": ["Adam", "Andrzej", "Anzelm", "Alojzy", "Arkadiusz"], "wiek": [19, 36, 94, 56, 42]})
tab_f = pd.DataFrame({"wiek": [36, 94, 42, 82], "sztuczna_szczeka": [False, True, False, True]})
print(tab_e)
kreski()
print(tab_f)
kreski()
pd.merge(tab_e, tab_f, on = "wiek", how = "right")
# Tym razem, wszystkie wyniki, które nie znalazły się w tabeli 2 zostały odrzucone. Mamy za to wartość z tabeli tab_f, która nie ma odpowiednika w tab_e.
```

```python
# Łączenie tabel typu OUTER JOIN:
tab_e = pd.DataFrame({"imie": ["Adam", "Andrzej", "Anzelm", "Alojzy", "Arkadiusz"], "wiek": [19, 36, 94, 56, 42]})
tab_f = pd.DataFrame({"wiek": [36, 94, 42, 82], "sztuczna_szczeka": [False, True, False, True]})
print(tab_e)
kreski()
print(tab_f)
kreski()
pd.merge(tab_e, tab_f, on = "wiek", how = "outer")
# Tym razem, mamy wszystkie elementy obu tabel.
```

```python
# Łączenie tabel typu INNER JOIN:
tab_e = pd.DataFrame({"imie": ["Adam", "Andrzej", "Anzelm", "Alojzy", "Arkadiusz"], "wiek": [19, 36, 94, 56, 42]})
tab_f = pd.DataFrame({"wiek": [36, 94, 42, 82], "sztuczna_szczeka": [False, True, False, True]})
print(tab_e)
kreski()
print(tab_f)
kreski()
pd.merge(tab_e, tab_f, on = "wiek", how = "inner")
# Tym razem, mamy jedynie elementy tabel, które mają wspólny klucz w postaci zmiennej "wiek".
```

```python
# Podstawowe obliczenia i operacje na DataFrame
# Uzupełnianie i usuwanie braków w danych:
dane = pd.read_csv("/content/lista_projektow.csv", delimiter = ";")
dane
```

```python
# Wyświetlenie podsumowania statystycznego na danych liczbowych:
dane["Wartość projektu [zł]"].describe()
```

```python
# A jeśli chcemy wyciągnąć konkretną wartość:
print("Liczba projektów: ", dane["Wartość projektu [zł]"].count())
print("Minimalna wartość: ", dane["Wartość projektu [zł]"].min())
print("Maksymalna wartość: ", dane["Wartość projektu [zł]"].max())
print("Średnia: ", dane["Wartość projektu [zł]"].mean())
```

```python
# Wyświetl unikalne elementy w kolumnie:
# W jakich programach (kategoriach) były realizowane projekty?
dane["Kategoria"].value_counts()
```

```python
# Liczba unikalnych elementów w kolumnie:
# Ilu beneficjentów skorzystało z dofinansowania?
dane["Nazwa beneficjenta"].nunique()
```

```python
# Usuwanie duplikatów wierszy:
dane.drop_duplicates()
```

```python
# Usuwanie duplikatów wierszy w wybranej kolumnie:
# Usuwany wszystkie wartości, w których powtarza się jakaś wartość z tej kolumny, z wyjątkiem pierwszej:
dane.drop_duplicates(subset = ["Tytuł projektu"], keep = "first")
```

```python
# Usuwanie duplikatów wierszy w wybranej kolumnie:
# Usuwany wszystkie wartości, w których powtarza się jakaś wartość z tej kolumny, z wyjątkiem ostatniej:
dane.drop_duplicates(subset = ["Tytuł projektu"], keep = "last")
```

```python
# Sortowanie i grupowanie wierszy.
# Sotrujemy w kolejności rosnącej wartości projektu:
dane.sort_values(by = "Wartość projektu [zł]", inplace = True, ascending = True)
dane
```

```python
# Grupowanie elementów w tabeli.

print("Przykład 1: Chcemy wiedzieć, ile wyniosła łączna suma dofinansowania w ramach poszczególnych kategorii:")
print(dane.groupby("Kategoria")["Dofinansowanie z EU [zł]"].sum())
kreski()

print("Przykład 2: Chcemy wiedzieć, jaka była średnia wartość projektów realizowanych w ramach jednej jednostki administracyjnej oraz w ramach wielu jednostek administracyjnych:")
print(dane.groupby("Projekt jest realizowany na obszarze większym niż jedna jednostka administracyjna?")["Wartość projektu [zł]"].mean())
kreski()

print("Przykład 3: Chcemy wiedzieć, ile warty był najdroższy projekt realizowany w ramach każdej kategorii:")
print(dane.groupby("Kategoria")["Wartość projektu [zł]"].max())
kreski()

print("Przykład 4: Chcemy wiedzieć, ile projektów zrealizował każdy z beneficjentów w ramach każdej kategorii:")
print(dane.groupby(["Kategoria", "Nazwa beneficjenta"])["Wartość projektu [zł]"].count())
```

```python
# A teraz statystyki dla interesujących nas danych z tabeli, np. całkowite wydatki klientów:
dane.groupby("Kategoria")["Wartość projektu [zł]"].describe()
```

```python
# Wartości brakujące w danych:
# W ramach PANDAS dostępne są 4 metody radzenia sobie z brakami danych:
# 1) isnull() - generuje maskę logiczną wskazującą brakujące wartości,
# 2) notnull() - przeciwnie do isnull,
# 3) dropna() - zwraca dane bez brakujących wartości,
# 4) fillna() - zwraca dane z wypełnionymi lub zamienionymi brakującymi wartościami
```

```python
# Oto nasza tabela:
import numpy as np

tabela = pd.DataFrame({"Państwo": ["Polska", "USA", "Mauritius", "Cypr", "Wenezuela", "Filipiny", "Australia", "Liechtenstein"],
                      "Ludność_mln": [38.0, 300.0, np.nan, 1.5, 15.0, 149.5, 34.0, np.nan],
                      "Morze": [True, True, True, True, True, np.nan, True, False],
                      "UE": [True, False, False, True, np.nan, False, False, False]})
tabela
```

```python
# Wyświetl wiersze, w których brakuje wartości w kolumnie "Ludność_mln":
tabela[tabela["Ludność_mln"].isnull()]
```

```python
# Usuwanie elementów z brakującymi wartościami:
tabela.dropna()

# Zostały usunięte wszystkie wiersze, w których zabrakło choćby jednej wartości
```

```python
# Usuwanie elementów z brakującymi wartościami w wybranej kolumnie - sposób 1
# Nie chcemy wyświetlać krajów, w których nie ma informacji o dostępie do morza:
tabela[tabela["Morze"].notna()]
```

```python
# Usuwanie elementów z brakującymi wartościami w wybranej kolumnie - sposób 2
# Nie chcemy wyświetlać krajów, w których nie ma informacji o dostępie do UE:
tabela[tabela["UE"].notnull()]
```

```python
# Usuwanie wierszy, w których znajduje się mniej niż 4 rzeczywiste próbki:
tabela.dropna(thresh = 4)

# Zniknęły wszystkie elementy, w których nie było co najmniej 4 niepustych kolumn.
```

```python
# Usuwanie kolumn z wartościami NAN:
tabela.dropna(axis = 1)
```

```python
# Usuwanie wierszy z wartościami NAN:
tabela.dropna(axis = 0)
```

```python
# Zamieniamy wartości brakujące na 0:
tabela.fillna(0)
```

```python
# Zamieniamy wartości brakujące na średnie:
tabela.fillna(tabela.mean())

# Wydaje mi się, że wartości powyżej 0,5 są trochę bardziej True a te poniżej - False :-P
```
